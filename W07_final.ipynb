# Summary


For this task we combined two datasets covering 
    1. The 2017 german primary election results
    2. Structural data about the areas that the votes are counted in (constituencies / areas). The structural data covers many different categories such as religion, income, birth rate, etc.

We then proceed to analyse the following:
    1. Which features are strongly associated with which party.
    2. Which areas did better and worse than expected for each party. 
    
These suggestions can be used by the parties to gain information about their voter types and areas that should be investigated further in order to explain the difference in expected and actual results.

In order to achieve these results we used different analysis methods learned throughout the course and added `MultiOutput Regression` (https://scikit-learn.org/stable/modules/generated/sklearn.multioutput.MultiOutputRegressor.html) which was not covered.

```python
# Data processing

## Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
import warnings
warnings.filterwarnings('ignore')

# Data Visualization

## Libraries
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
## options
pd.set_option('display.max_rows', 20)

# Regression

## Libraries
from sklearn.feature_selection import f_regression
from sklearn.feature_selection import SelectKBest
```

### Import the data

```python
# We import the results dataset into python.
results = pd.read_csv("results_transformed.csv", sep=";", decimal=',')

# List of all party's names
parties = list(results.columns[7:])
parties
```

```python
# We import the features dataset into python.
features = pd.read_csv("features_transformed.csv", sep=";", decimal=',')
```

### A little further data proccesing to get everything up & running.

```python
# for some reason decimal = "," doesn't get rid of all commas in the excel file so we need to transform the rest
col2replace = list(features.select_dtypes(exclude=['float64', 'int64']).columns)

# state is non-numerical so does not need to be transformed from european , to us . notation
col2replace.remove('state')

# some NAN valued are in the table (indicated by a . with no value so we put nan there)
features[col2replace] = features[col2replace].replace(".", np.nan)

# replace all, by a dot
features[col2replace] = features[col2replace].apply(lambda x: pd.to_numeric(x.str.replace(",",".")))
```

```python
# this should return 0 indicating that all commas have been transformed to dots in order to get them to numerical data types
features[features.eq(",").any(1)].shape[0]
```

```python
# Merge the results dataframe with the feature dataframe using merge
df_all = results.merge(features, left_on='constituency number',right_on='constituency number')
df_all
```

```python
# Checking if all columns are there
# -1 because constituency number is in both -> will only be included once
df_all.shape[1] == features.shape[1] + results.shape[1] - 1
```

### Changing all to proportional data

```python
# get voters as a percentage of all people in that area that were eligible to vote
df_all["voters %"] = df_all["voters"] / df_all["eligible voters"].astype(float) * 100

# get invalid votes as a percentage of the votes casted
df_all["invalid votes %"] = df_all["invalid votes"] / df_all["voters"].astype(float) * 100

# get valid votes as a percentage of votes casted
df_all["valid votes %"] = df_all["valid votes"] / df_all["voters"].astype(float) * 100

# get number of votes for other parties
df_all["other parties"] = df_all["valid votes"] - (df_all.loc[:, ["cdu", "spd", "linke", "fdp", "afd", "green"]].sum(axis=1))
# get percentage of valid votes that went to other parties than the ones who got into parliament
df_all["other parties %"] = df_all["other parties"] / df_all["valid votes"].astype(float) * 100

# get percentages of valid votes that went to each specific party
df_all["cdu %"] = df_all["cdu"] / df_all["valid votes"].astype(float) * 100
df_all["spd %"] = df_all["spd"] / df_all["valid votes"].astype(float) * 100
df_all["linke %"] = df_all["linke"] / df_all["valid votes"].astype(float) * 100
df_all["fdp %"] = df_all["fdp"] / df_all["valid votes"].astype(float) * 100
df_all["afd %"] = df_all["afd"] / df_all["valid votes"].astype(float) * 100
df_all["green %"] = df_all["green"] / df_all["valid votes"].astype(float) * 100
```

```python
# Checking if all votes were distributed correctly as a percentages we check if everything adds to 100%
df_all.loc[:, ["cdu %", "spd %", "linke %", "fdp %", "afd %", "green %", "other parties %"]].sum(axis=1)
```

### Separate in Country, State and Area-wise information.

```python
# seperate areas, states and germany as whole to each own dataframe

# get all areas
areas = df_all[df_all["constituency number"] < 900]

# get all states
states = df_all[np.logical_and(df_all["constituency number"] > 900, df_all["constituency number"] < 999)]

# get germany as a whole
germany = df_all[df_all["constituency number"] == 999]
```

```python
# parties results as % column header names
parties_percentages = ["cdu %", "spd %", "linke %", "fdp %", "afd %", "green %", "other parties %"]
```

```python
# Compute the correlation matrix
corr = areas.corr()
corr = corr[parties_percentages]
corr
```

```python
# Copy DF
X = areas.copy()
# We drop metadata and target variables columns
X = X.drop(columns=['constituency number', 'belongs to', "eligible voters", "voters", "invalid votes", \
                    "valid votes", "cdu", "spd", "linke", "green", "fdp", "afd", \
                    "voters %", "invalid votes %", "valid votes %", "cdu %", "spd %", "linke %", "fdp %", \
                    "afd %", "green %", "other parties", "other parties %", "area name", "state"])
# FillNA with value 0
X = X.fillna(0)
```

```python
# Define target variable
y = areas[parties_percentages]
y
```

### Building a Regression Model

```python
# Create a train and test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# In order to test for each party run cross validation on each party prediction
# we do this because cross validation is not supported for multi-output regressions so we run our own version of it here
# by fitting a regressor for each party and then comparing the results
for party in y.columns:
    clf = Ridge(alpha=10)
    scores = cross_val_score(clf, X_train, y_train[party].values, cv=5, scoring='r2')
    print("------")
    print(party)
    print(round(scores.mean(),2))
```

```python
# we continue with multi-output regression because it provides a nice wrapper around the functionality
clf = MultiOutputRegressor(Ridge(alpha=10,random_state=42)).fit(X_train, y_train)
```

```python
# Average r2 scores of accross parties for train and test
print("Train Score: ",clf.score(X_train, y_train))
print("Test Score: ",clf.score(X_test, y_test))
```

```python
# In order to find constituencies that have performed very different from what would have anticipated we will fit on the entire data
# then we "predict" on the entire dataset and compare the the results with the real percentages the parties received
# that we we find values that are far away from our regression line and might be worth looking at from a business perspective
clf = MultiOutputRegressor(Ridge(alpha=10,random_state=42)).fit(X, y)
y_pred = clf.predict(X)
```

```python
for i, party in enumerate(parties):
    areas.loc[:, party + " pred %"] = y_pred[:,i].reshape(-1,1)
```

```python
for party in parties:
    diff_cdu = pd.DataFrame(areas[party + " %"] - areas[party + " pred %"], columns = ["diff"])
    diff_cdu["area name"] = areas["area name"]
    print("\n ------------ " + party.upper() + " ------------\n")
    print("TOP 3 OverPerformed Areas")
    display(diff_cdu.sort_values(by=["diff"], ascending=False).head(3))
    print("\nTOP 3 UnderPerformed Areas")
    display(diff_cdu.sort_values(by=["diff"], ascending=False).tail(3))
```

## Lasso Regression for selecting variables

With Lasso Regression a penalty is applied for high amount of variables, so setting a relatively high $\alpha$ will take most of the variable's coefficients to zero, then we will get the most important variables for explaining the variability of each party's percentage amount of votes per area.

```python
# Custom function for Lasso regression
# t = regression type
# r = range of indizes of features to include
# kwargs = hyperparameters for regression

def LassoRegression(X_train, y_train, r, party, **kwargs):

    X = X_train.iloc[:, r].values
    y = y_train
    
    alpha = kwargs.get('alpha',1)
    print(f"ALPHA {alpha}")
    reg = Lasso(alpha=alpha).fit(X, y)
    print(reg)

    coef = pd.DataFrame(reg.coef_, index=list([X_train.columns[r]]), columns=[party])
    print(f"Train score: {reg.score(X, y)}")
    display(coef[coef.values !=0].sort_values(by=[party], ascending=False))
```

## Best Subset

This last option is similar to Lasso because it will leave us the best variables for explaining the variability of the target variables, but the evaluation it's made based on the `p-value` of a hypothesis test wether a variable coefficient of a model regression is or it isn't zero.

```python
## This 
def best_subset(party):
    print("\n------------------------------------------------------------------------\n BestSubset")
    print(" {}".format(party.upper()))
    aux = SelectKBest(score_func=f_regression, k=10).fit(X, y[party + " %"])
    display(pd.DataFrame(X.columns[list(pd.Series(aux.pvalues_).sort_values().head(10).index)], columns=[party]))
```

```python
def full_pipeline(X, y, party):
    __cor(party)
    print("------------------------------------------------------------------------ \n LASSO")
    LassoRegression(X, y[party + " %"], range(len(X.columns)), party, alpha=10)
    best_subset(party)
    
```
